### Reduction

Calculating the sum across all elements of a tensor is a common operation. For example:

* For vector ***x*** of length *n*, we calculate $\sum_{i=1}^{n} x_i$
* For matrix ***X*** with *m* by *n* dimensions, we calculate $\sum_{i=1}^{m} \sum_{j=1}^{n} X_{i,j}$

### The Dot Product
If we have two vectors (say, ***x*** and ***y***) with the same length *n*, we can calculate the dot product between them. This is annotated several different ways, including the following:

* $x \cdot y$
* $x^Ty$
* $\langle x,y \rangle$

We calculate products in an element-wise fashion and then sum reductively across the products to a scalar value. That is, $x \cdot y = \sum_{i=1}^{n} x_i y_i$

Dot product between $x$, $y$ can be interpreted as projecting $x$ onto $y$ (or vice-versa), which measures the similarity between the two vectors.

```python
x = np.array([25, 2, 5])
y = np.array([0, 1, 2])
```

```python
25*0 + 2*1 + 5*2
```

```python
np.dot(x, y)
```





### Convolution

Convolution is a mathematic operation that can be performed on two signals, which produces a third output signal which is a “blend” of the two inputs.

$(x * y)[n] = \sum_{m= 0}^{N-1}  x[m]g[n-m]$


Physically, It may be interpreted as similarity test. If we are convoluting two signals, we shift and compute dot product. If they are different, result is random, but when on particular shift they expose similarity the result will be positive and the more similarities the higher is.

corr.gif

Which can be easily generalized to 2-D case.

conv_similarity.gif

Convolution can be also interpreted as taking the signal and replacing each point by the weighted average of it and its two neighboring points, which probably makes the signal change more slowly, because any big changes will get smoothed out by averaging them with nearby points.



corr_2.gif



```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 1, 1])
y = np.array([1, 1, 1])/3

z = np.convolve(x,y)
```

```python
fig, axs = plt.subplots(1, 3,figsize=(20,5))



axs[0].set_title('Signal x', fontsize=16)
axs[1].set_title('Signal y', fontsize=16)
axs[2].set_title('Convolution Result', fontsize=16)

axs[0].set_xlim([-5, 5])
axs[1].set_xlim([-5, 5])
axs[2].set_xlim([-5, 5])


axs[0].scatter(range(len(x)),x)
axs[1].scatter(range(len(y)),y)
axs[2].scatter(range(-int(len(z)/2),int(len(z)/2)+1),z)
```

Since, noise are mostly a random fastly changing signal added, it can be minimized using convolution.

```python
n = np.arange(0, 100, 0.1)
x = np.sin(n) + np.random.normal(0, 0.1, len(n))
filter = np.array([1, 1, 1 , 1 , 1])/5


filtered_x = np.convolve(filter,x)
```

```python
fig, axs = plt.subplots(1, 2,figsize=(20,5))


axs[0].set_title('Original Signal', fontsize=16)
axs[1].set_title('Filtered Signal', fontsize=16)

axs[0].set_xlim([-3, 100])
axs[1].set_xlim([-3, 100])

axs[0].scatter(range(len(x)),x)
axs[1].scatter(range(len(filtered_x)),filtered_x)
```

These pixels contain information about color and intensity. Here's an example of the matrix for a 2D grayscale image. Here we can see that the first image is a pixelated image. The previous example represented a Grayscale image which only has shades of black and white. To obtain colored image, 2-dimensional color images are often represented in RGB—3 layers of 2-dimensional arrays, where the three layers represent Red, Green and Blue channels of the image

```python
#Import cv2, which is an open source computer vision and machine learning software library
import cv2
```

```python
img = cv2.imread("image.png") # reads an image
print(img.shape) # print the 3 dimentional colored image size
plt.imshow(img) # displays the image "img"
```

```python
(height, width,channels) = img.shape
print (height, width, channels)
```

```python
#Display the content of red channel only (2D)
red = img[:,:,2]
plt.imshow(red,cmap='gray')
```

```python
#Display the content of blue channel only (2D)
blue = img[:,:,0]
plt.imshow(blue,cmap='gray')
```

```python
#Reads the image as greyscale
grey = cv2.imread("image.png",0)
print(grey.shape)
plt.imshow(grey,cmap='gray')
```

```python
#Another way to convert RGB to greyscale
grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(grey.shape)
plt.imshow(grey,cmap='gray')
```

```python
#slicing the image
img2=img[0:150,0:150]
plt.imshow(img2)
```
